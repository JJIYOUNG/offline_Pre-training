#### 1. 반갑다 주인! 나는 앵무새!
지난시간에 print() 로 출력을 배웠다면, 이번 시간에는 입력을 다뤄봅시다.

입력은 컴퓨터에게 정보를 전달해주는 것을 의미합니다. 기존에 우리가 코드를 작성하여 컴퓨터에게 '명령'만을 전달했다면, 이제는 이에 '추가정보'를 주어 그 정보를 이용하여 명령을 수행할 수 있습니다.

우리가 말하면 그대로 따라서 말하는 앵무새처럼
입력과 출력을 이용해서 우리가 입력한 값을 그대로 출력하는 앵무새 프로그램을 만들어봅시다.

```python
# input()을 이용해서 입력을 넣고, 이를 변수 var에 담아봅시다.
var = input()

# 앵무새가 말을 따라합니다!
print("앵무새 :", var)

# 입력한 값이 어떤 자료형인지 확인해봅시다.
print(type(var))
```

#### 2. input()을 이용해서 숫자(정수) 입력을 받고, 변수 num에 이를 넣어봅시다.

if-else문을 이용해서 만약 입력받은 수가 홀수면 "(입력받은 수) 홀수입니다."<br>
짝수면 "(입력받은 수) 짝수입니다."를 출력해봅시다.
괄호는 출력하지 않습니다.

ex) 입력값 5<br>
출력결과 >>  5는 홀수입니다.

```python
num = int(input("정수를 입력하세요>> "))

# 홀수 or 짝수
if num%2 != 0:
  print(f'{num} 홀수입니다.')
else:
  print(f'{num} 짝수입니다.')
```

#### 3. 한 자리수, 두 자리수, 세 자리수! 구분할 수 있나요?
1~999까지의 숫자 중 하나가 입력될 때,

- 1자리(1~9)이면 "한 자리 숫자입니다."
- 2자리(10~99)이면 "두 자리 숫자입니다."
- 3자리(100~999)이면 "세 자리 숫자입니다."

```python
# 변수 num을 선언하고, 숫자형으로 입력을 받습니다.
num = int(input("정수로 입력하세요(1~999)>> "))

# if-elif-else문을 이용해서 조건에 따라 출력합니다.
# 왼쪽에 있는 조건에 따라 자리수를 출력해봅시다.
if num >= 100:
  print("세 자리 숫자입니다.")
elif num >= 10:
  print("두 자리 숫자입니다.")
else:
  print("한 자리 숫자입니다.")
```

#### 7. 1~50까지의 범위에서 랜덤 숫자 하나를 뽑아 com_num에 담고, input을 통해 숫자형으로 입력받아 user_num 변수에 저장, UP-Down Game을 만들어보자.

- 만약 com_num보다 user_num이 더 크면 "{입력값} 보다 작은 숫자 입니다."
- 만약 com_num보다 user_num이 더 작으면 "{입력값} 보다 더 큰 숫자입니다."
- 만약 com_num과 user_num이 같으면 "{입력값} 정답! 게임을 종료합니다." 를 출력합니다. 정답을 출력하면 게임을 종료합니다!

```python
# 1~ 100 범위에서 랜덤수 출력하는 방법
import random
random.randint(1,100)
```

```python
com_num = random.randint(1,50)
while True:
  user_num = int(input("입력>> "))
  if com_num < user_num:
    print(f"{user_num}보다 더 작은 숫자입니다.")
  elif com_num > user_num:
    print(f"{user_num}보다 더 큰 숫자입니다.")
  else:
    print(f"{user_num} 정답! 게임을 종료합니다.")
    break
```

#### 10. 5명에 대한 정보처리기사 자격증 시험 점수가 리스트에 담겨있습니다. 이때 각 점수가 합격 점수인지 불합격 점수인지 판별하여 출력하시오. (60점 이상 합격)
- 조건: range() 사용 <br>
- score_list = [90,45,70,60,55]<br>
- 출력 결과 >>
  - 0번 학생은 합격입니다.
  - 0번 학생은 불합격입니다.

```python
score_list = [90,45,70,60,55]
for i in range(len(score_list)):
    if score_list[i] >= 60:
        print(f"{i+1}번 학생은 합격입니다.")
    else:
        print(f"{i+1}번 학생은 불합격입니다.")
```

#### 12. 2단과 3단을 외워보자!

2 * 1 = 2　　　3 * 1 = 3<br>

#### 13. 이중 리스트에 학생 이름과 수학 성적 데이터가 담겨있다. 학생들의 수학 점수의 평균은 얼마인가?

- score_data = [['연재',80],['고은',55],['형식',75],['민채',95],['모은',100]]
- 출력결과 >>
  - 수학 점수 평균: 81.0

```python
score_data = [['연재',80],['고은',55],['형식',75],['민채',95],['모은',100]]
total = 0
for i in score_data:
  total += i[1]
avg = total/len(score_data)
print(f'수학 점수 평균: {avg}')
```
