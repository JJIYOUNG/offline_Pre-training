### 함수 vs 메서드
- 함수(function): 특정한 작업을 수행하는 코드 집합
- 메서드(method): 특정 자료에 대해 특정 기능을 하는 코드 집합

```python
# 내장 함수
sum([1,2,3])
print('안녕하세요!')
len([1,2])

# 문자열 메서드
str1 = '안녕하세요!'
print(str1.count('!'))
print(str1.index('!'))

# 리스트 메서드
lst1 = [1,2,3,3,3]
print(lst1.count(3))
lst1.reverse()
print(lst1)
```

#### 3. BMI(체질량지수) 계산하기
BMI(체질량지수)는 우리 몸의 신장과 몸무게를 통해 계산되는 수치로, 비만도를 판단하는데 사용됩니다.

BMI 수치를 계산하는 방법은 다음과 같습니다.
BMI = 체중(kg) / (신장(m)의 제곱)

마지막으로 BMI의 수치에 따라 4가지 등급으로 나누어집니다.

- 30 |  비만
- 25 ~ 29 | 과체중
- 20 ~ 24 | 정상
- 20 미만 | 저체중

신장(cm)과 체중(kg) 데이터가 있는 딕셔너리를 인자로 받아 BMI 수치를 알려주는 BMI 함수를 만들어보겠습니다.

- body_data를 그대로 함수에 인자로 사용
- 함수 실행 결과>> BMI : 24.16

```python
# BMI를 계산하는 함수 BMI()를 만들어봅시다.
# BMI를 계산할 때, 사용하는 신장의 단위는 m(미터) 입니다
def BMI(body_info) :
  bmi = body_info['몸무게'] / ((body_info['신장'] * 0.01)**2)
  return round(bmi,2)

# 저장된 데이터의 신장의 단위는 cm, 몸무게의 단위는 kg입니다.
body_data = {
    "신장" : 175,
    "몸무게" : 74
}

print("BMI : ", BMI(body_data))

# 제곱 함수
# import math
# math.pow(2,3) 2의 3승
```

#### 4. Person 클래스를 생성하시오. '지영',20 정보를 가진 sjy객체를 생성하고 고은 이라는 친구 이름을 부르며 인사하기 !!
(단, 객체를 생성할 때 이름과 나이를 입력받으시오.)

- 속성: 이름, 나이
- 메서드:
  - 먹기: "냠냠!!" <br>
  - 인사하기: "나는 지영이야. 고은 만나서 반가워!"
    * 친구 이름을 입력받아 이름을 부르며 인사하기 <br>


```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def eat(self):
    return "냠냠!!"

  def hi(self, fr_name):
    return f'나는 {self.name}이야. {fr_name} 만나서 반가워!'
```

```python
# 고은 이라는 친구 이름을 부르며 인사하기 !!

sjy = Person('지영',20)
sjy.hi('고은')
```

#### 5. Person 클래스를 상속받은 Person2 자식 클래스를 만들고, 부모 클래스의 eat 메서드를 오버라이딩하시오. 원호 , 30 속성을 가진 pwh객체 만들고 eat메서드를 실행하자!

- eat: 냠냠!! 와구와구!!

```python
class Person2(Person):
  def eat(self):
    print(super().eat(), end = ' ')
    print("와구와구!!")
```

```python
# 원호 , 30 속성을 가진 pwh객체 만들고 eat메서드를 실행하자!
pwh = Person2('원호',30)
pwh.eat()
```
