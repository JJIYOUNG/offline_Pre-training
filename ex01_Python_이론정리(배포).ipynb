#### 1. 파이썬? 그게뭐야 먹는거야?

#### 2. 변수와 데이터타입
- 변수 : 데이터 공간에 이름표를 붙여보자
- 단일형/ 시퀀스형 데이터타입 : 자전거와 기차를 생각해봐
- 단일형 데이터타입 : 숫자(int, float), 문자(character), 불리언(True, False)
- 시퀀스형 데이터타입: 문자열(str), 리스트(list), 튜플(tuple), 딕셔너리(dict), ...

#### 3. 연산자
- 산술연산자: +,-,/,*,//,%
- 비교연산자: 같다 같지않다, 크다 작다를 표현하지

  - count()
  - find(), index()
  - replace()
  - strip(), lstrip(), rstrip()
  - split()
  - upper(), lower()

- 연산 : 문자열 더하기가 가능해!

#### 5. 리스트와 튜플
- 리스트 : 여러 값들을 줄세워 놓는 리스트, 새로운 값 받을 수 있어!

- 튜플 : 더이상 새로운 값을 받을 수 없어! 들어있는 값 확인만 가능하지!

#### 6. 딕셔너리와 집합({1,3,2})
- 딕셔너리(dict) : 말그대로 사전, 단어와 뜻이있는 짝꿍

- 집합 특징: 중복허용x, 순서가 없지(Unordered)
- 집합 사용: set은 중복을 허용하지 않는 특징때문에 중복 제거를 위한 필터로 종종 사용해!

#### 7. 제어문

#### 8. 함수
- 함수를 쓰는 이유 : 다이어트를 위해 매일 똑같은 운동 루틴을 실행할거야! 계속 같은 기능을 불러다 써야할 때

#### 3. ValueError 함수안에 인자를 잘못넣은경우

```python
# int('123')
int('hello world')
# int는 정수형을 넣어줘야하는데 문자열이 들어가서 생기는 오류
```

#### 4.TypeError 자료형이 맞지않을때

```python
x = '보라색'
y = '파랑색'
z = x - y
print(z)

# 자료형이 맞지 않아서 생기는 오류, 문자열이라 빼기를 할 수 없음
```

#### 5.ZeroDivisionError 0으로 나눌수 없는 오류

```python
x = 3
y = 0
z = x/y
print(z)

# 0으로 나눌수 없는데 나눠서 생기는 오류
```
