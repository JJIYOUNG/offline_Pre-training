### 함수 vs 메서드
- 함수(function): 특정한 작업을 수행하는 코드 집합
- 메서드(method): 특정 자료에 대해 특정 기능을 하는 코드 집합

```python
# 내장 함수
print(sum([1,2,3]))
print('안녕하세요!')
print(len([1,2]))

# 문자열 메서드
str1 = '안녕하세요!'
print(str1.count('!'))
print(str1.index('!'))

# 리스트 메서드
lst1 = [1,2,3,3,3]
print(lst1.count(3))
lst1.reverse()
print(lst1)
```
